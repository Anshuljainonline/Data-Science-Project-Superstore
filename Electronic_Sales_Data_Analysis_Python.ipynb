### Read all data file

```python
all_data= pd.read_csv("all_data.csv")
all_data.head()
```

## Task 2: Removing NaN

```python
all_data=all_data.dropna()
```

## Task 3: Find Or and Delete it

```python
temp_df= all_data[all_data["Order Date"].str[0:2]=='Or']
temp_df
```

```python
all_data2= all_data[all_data.Month.str[0:2] != 'Or']
all_data2.head()
```

## Task 4: Convert Column to correct type

```python
all_data2["Quantity Ordered"]= pd.to_numeric(all_data2["Quantity Ordered"])
all_data2["Price Each"]= pd.to_numeric(all_data2["Price Each"])
```

## Task 5: Adding more columns

## 5.1 Adding Month Column

```python
all_data2["Month"]=all_data2["Order Date"].str[0:2]
all_data2["Month"]= all_data2.Month.astype("int32")
```

```python
all_data2.head()
```

## 5.2 Add Sales Column

```python
all_data.
```

```python
all_data2["Sales"]= all_data2["Quantity Ordered"] * all_data2["Price Each"]
```

```python
all_data2.head()
```

## 5.3 Add city column

```python
all_data2["City"]=all_data2["Purchase Address"].apply(lambda x: x.split(',')[1] + " (" + x.split(",")[2].split(" ")[1] +")" )       

all_data2.head()
```

## Analysis 1: Best Month For Sales

```python
results= all_data2.groupby("Month").sum()
results.head()
```

```python
plt.bar(results.index, results.Sales, color='red')
plt.xticks(results.index)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

## Analysis 2: Wat city has highest number of sales?

```python
result2=all_data2.groupby("City").sum()
result2
```

```python
plt.bar(result2.index,result2.Sales, color="yellow")
plt.xticks(result2.index,rotation='vertical')
plt.show()
```

## Analysis 4: What products are sold together?

```python
df =all_data2[all_data2["Order ID"].duplicated(keep =False)]
```

```python
df.head(7)
```

```python
df["Grouped"]= df.groupby("Order ID").Product.transform(lambda x: ','.join(x))

df=df[["Order ID", "Grouped"]].drop_duplicates()

df.head(30)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df.Grouped:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
count.most_common(10)
```
